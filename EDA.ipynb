```python
for column in data.columns:
    if data[str(column)].isna().sum() > 0.3 * len(data):
        data.drop(columns=str(column), inplace=True)

data.isna().sum()
```

```python
data.drop(columns= ["ID", "VFN", "Mp", "Mh", "Man",  "Tan", "T", "Va", "Ve", "Ct", "Cr", "r"], inplace=True)
```

```python
data.head()
```

```python
data.drop(['Mt','year','Status'],axis=1,inplace=True)
```

```python
mapper = {'Mk': 'Make',
          'Cn': 'Model',
          'm (kg)': 'Weight',
          'Ewltp (g/km)': 'CO2 Emissions',
          'Ft': 'FuelType',
          'Fm': 'FuelMode',
          'ec (cm3)': 'EngineSize',
          'ep (KW)':'EnginePower',
          }
data.rename(columns=mapper,inplace=True)
```

```python
data.isna().sum()
```

```python
data['EnginePower'] = pd.to_numeric(data.EnginePower*1.341)
```

```python
data[data["EnginePower"].isnull()]
```

```python
data = data[(data['Make'].notna()) & (data['Model'].notna())]
```

```python
data['Date of registration']= pd.to_datetime(data['Date of registration'])
```

```python
data['EnginePower'] = [int(i) if i > 0 else i for i in data['EnginePower']]
```
